```python
import os
import glob
import pandas as pd
import numpy as np
from onekey_algo.custom.components.Radiology import ConventionalRadiomics
import pickle
from sklearn.preprocessing import StandardScaler
```

```python
def pred_by_model(model,data):
    with open(model, 'rb') as f:
        clf = pickle.load(f)
    sd = StandardScaler()
    data = pd.DataFrame(sd.fit_transform(data),index=data.index,columns=data.columns)
    pred = pd.DataFrame(clf.predict_proba(data),index=data.index)
    return pred
```

```python
def Feature_extra(images,masks):
    radiomics = ConventionalRadiomics(param_file)
    radiomics.extract(images, masks)
    rad_data = radiomics.get_label_data_frame(label=1)
    return rad_data
```

```python
data_root = r''
param_file = './exampleCT.yaml'
```

```python
left_images = sorted(glob.glob(os.path.join(data_root, 'left','images', '*.nii.gz')))
left_masks = sorted(glob.glob(os.path.join(data_root,'left','masks','*.nii.gz')))
```

```python
right_images = sorted(glob.glob(os.path.join(data_root, 'right','images', '*.nii.gz')))
right_masks = sorted(glob.glob(os.path.join(data_root,'right','masks','*.nii.gz')))
```

```python
left_data = Feature_extra(left_images,left_masks)
right_data = Feature_extra(right_images,right_masks)
```

```python
left_features = pd.read_table('left.txt',header=None).iloc[:,0].to_list()
right_features = pd.read_table('right.txt',header=None).iloc[:,0].to_list()
# left_data[left_features].to_excel('left.xlsx')
# right_data[right_features].to_excel('right.xlsx')
```

```python
left_data = left_data[left_features]
right_data= right_data[right_features]
```

```python
left = pred_by_model('leftRandomForest.pickle',left_data)
right = pred_by_model('rightRandomForest.pickle',right_data)
```

```python
left['Pred']=None
right['Pred']=None
```

```python
left.loc[left.iloc[:,1]>=0.38,'Pred'] = 1
left.loc[left.iloc[:,1]<0.38,'Pred'] = 0
right.loc[right.iloc[:,1]>=0.52,'Pred'] = 1
right.loc[right.iloc[:,1]<0.52,'Pred'] = 0
```

```python
left.to_excel('predleft.xlsx')
right.to_excel('predright.xlsx')
```
